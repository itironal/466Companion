```python
#### Sampling example
import matplotlib.pyplot as plt
from skimage import data, color, io
from skimage.transform import rescale, resize

# Load image from an external URL and save locally
!wget -cq https://raw.githubusercontent.com/itironal/466Companion/main/images/redmushroom.png

# Read the image from the saved file
image = io.imread('redmushroom.jpeg')

# Crop a specific region of the image (rows 800 to 1311, columns 513 to 1024)
image = image[800:1311, 513:1024]

# Calculate the pixel-to-inch ratio based on the figure's DPI (dots per inch)
px = 1 / plt.rcParams['figure.dpi']  # pixel in inches

# Downscale the image to different resolutions without anti-aliasing
# Rescale to 256 x 256 resolution
image_rescaled_256_256 = rescale(image, 1/2, anti_aliasing=False, channel_axis=2)

# Rescale to 128 x 128 resolution
image_rescaled_128_128 = rescale(image, 1/4, anti_aliasing=False, channel_axis=2)

# Rescale to 64 x 64 resolution
image_rescaled_64_64 = rescale(image, 1/8, anti_aliasing=False, channel_axis=2)

# Rescale to 32 x 32 resolution
image_rescaled_32_32 = rescale(image, 1/16, anti_aliasing=False, channel_axis=2)

# Rescale to 16 x 16 resolution
image_rescaled_16_16 = rescale(image, 1/32, anti_aliasing=False, channel_axis=2)

# Set up a figure with 2 rows and 3 columns of subplots, adjusting for DPI
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(3*512*px, 2*512*px))

# Flatten the 2D array of axes for easier indexing
ax = axes.ravel()

# Display the original image in the first subplot
ax[0].imshow(image)
ax[0].set_title("Original image", fontsize=20)
ax[0].axis('off')  # Hide axis ticks

# Display the 256 x 256 rescaled image
ax[1].imshow(image_rescaled_256_256)
ax[1].set_title("256 x 256 image", fontsize=20)
ax[1].axis('off')

# Display the 128 x 128 rescaled image
ax[2].imshow(image_rescaled_128_128)
ax[2].set_title("128 x 128 image", fontsize=20)
ax[2].axis('off')

# Display the 64 x 64 rescaled image
ax[3].imshow(image_rescaled_64_64)
ax[3].set_title("64 x 64 image", fontsize=20)
ax[3].axis('off')

# Display the 32 x 32 rescaled image
ax[4].imshow(image_rescaled_32_32)
ax[4].set_title("32 x 32 image", fontsize=20)
ax[4].axis('off')

# Display the 16 x 16 rescaled image
ax[5].imshow(image_rescaled_16_16)
ax[5].set_title("16 x 16 image", fontsize=20)
ax[5].axis('off')
```

```python
# Image interpolation: Upscale a low-resolution image (32 x 32) using different interpolation methods

# Apply nearest neighbor interpolation to upscale the image by a factor of 8
image_nn = rescale(image_rescaled_32_32, 8, anti_aliasing=False, channel_axis=2, order=0)

# Apply bilinear interpolation to upscale the image by a factor of 8
image_bilinear = rescale(image_rescaled_32_32, 8, anti_aliasing=False, channel_axis=2, order=1)

# Apply bicubic interpolation to upscale the image by a factor of 8
image_bicubic = rescale(image_rescaled_32_32, 8, anti_aliasing=False, channel_axis=2, order=3)

# Set up a figure with 1 row and 3 columns to display the images
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(3*512*px, 1*512*px))
ax = axes.ravel()

# Visualize the image upscaled using nearest neighbor interpolation
ax[0].imshow(image_nn, cmap='gray')
ax[0].set_title("Nearest neighbor interpolation", fontsize=14)
ax[0].axis('off')  # Hide axis ticks

# Visualize the image upscaled using bilinear interpolation
ax[1].imshow(image_bilinear, cmap='gray')
ax[1].set_title("Bilinear interpolation", fontsize=14)
ax[1].axis('off')

# Visualize the image upscaled using bicubic interpolation
ax[2].imshow(image_bicubic, cmap='gray')
ax[2].set_title("Bicubic interpolation", fontsize=14)
ax[2].axis('off')
```

## **Programming exercise 3.4:**

1. Run the code below to understand image transformations through resizing, rotation, and cropping.

```python
import numpy as np
import matplotlib.image
import matplotlib.pyplot as plt
import math
import urllib.request
from PIL import Image

# Function to calculate the rotated position for a given (x, y) coordinate.
# Takes in coordinates (x, y), the origin of rotation (o), and the rotation degree.
def find_rotated_pos(x: float, y: float, o, degree: float = 0):
    sin_val = math.sin((degree * math.pi) / 180)  # Calculate sine of the rotation angle
    cos_val = math.cos((degree * math.pi) / 180)  # Calculate cosine of the rotation angle
    t_x = x - o[0]  # Translate x-coordinate to origin
    t_y = y - o[1]  # Translate y-coordinate to origin
    # Rotate the coordinates
    r_x = t_x * cos_val - t_y * sin_val
    r_y = t_x * sin_val + t_y * cos_val
    # Translate rotated coordinates back to the original origin
    return r_x + o[0], r_y + o[1]

# Function to rotate an image by a specified degree using linear interpolation
def rotate_image_linear(img_func: np.ndarray, degree: float = 45) -> np.ndarray:
    shape = img_func.shape
    originPoint = (float(shape[0]) / 2.0, float(shape[1]) / 2.0)  # Set rotation origin at image center
    ret_val = np.zeros(shape)  # Initialize an array for the rotated image
    
    # Iterate through each pixel in the output image to set its new value based on rotation
    for x in range(0, shape[0]):
        for y in range(0, shape[1]):
            x_coor, y_coor = find_rotated_pos(float(x), float(y), originPoint, degree)
            if 0 <= x_coor <= shape[0] - 1 and 0 <= y_coor <= shape[1] - 1:
                x_floor = math.floor(x_coor)  # Get lower bound x-coordinate
                y_floor = math.floor(y_coor)  # Get lower bound y-coordinate
                x_ceil = math.ceil(x_coor)    # Get upper bound x-coordinate
                y_ceil = math.ceil(y_coor)    # Get upper bound y-coordinate
                
                # Apply bilinear interpolation based on surrounding pixels
                if x_floor == x_ceil and y_floor == y_ceil:
                    ret_val[x][y] = img_func[x_floor][y_floor]
                elif x_floor == x_ceil:
                    ret_val[x][y] = (
                        img_func[x_floor][y_floor] * (y_ceil - y_coor) +
                        img_func[x_floor][y_ceil] * (y_coor - y_floor))
                elif y_floor == y_ceil:
                    ret_val[x][y] = (
                        img_func[x_floor][y_floor] * (x_ceil - x_coor) +
                        img_func[x_ceil][y_floor] * (x_coor - x_floor))
                else:
                    ret_val[x][y] = (
                        img_func[x_floor][y_floor] * (x_ceil - x_coor) * (y_ceil - y_coor) +
                        img_func[x_floor][y_ceil] * (x_ceil - x_coor) * (y_coor - y_floor) +
                        img_func[x_ceil][y_floor] * (x_coor - x_floor) * (y_ceil - y_coor) +
                        img_func[x_ceil][y_ceil] * (x_coor - x_floor) * (y_coor - y_floor))

    return ret_val

# Set up a figure with 4 side-by-side subplots for visualizing each transformation step
fig, axs = plt.subplots(1,4,figsize=(15,6), gridspec_kw={'width_ratios': [1, 2, 2, 1]})
plt.rcParams.update({'font.size': 18})

# Read an image from a URL and convert it to a numpy array
url = "https://raw.githubusercontent.com/hazalmogultay/466Images/cedbb152431d662ceb613bef20238b829471b423/W3/interpolation_source_a1.png"
original_img = np.array(Image.open(urllib.request.urlopen(url)))

# Display the original image
axs[0].imshow(original_img)
axs[0].set_title('Original image', y=-0.3, fontsize=14)
axs[0].axis('off')

# Double the size of the original image by padding edges
shape = original_img.shape
double_sized_img = np.pad(original_img, ((shape[1]//2, shape[1]//2), (shape[0]//2, shape[0]//2), (0,0)), 'edge')

# Display the double-sized image
axs[1].imshow(double_sized_img)
axs[1].set_title('Double sized image', y=-0.15, fontsize=14)
axs[1].axis('off')

# Rotate the double-sized image by 45 degrees using linear interpolation
rotated_img = rotate_image_linear(double_sized_img, 45).astype(np.uint8)

# Display the rotated image
axs[2].imshow(rotated_img)
axs[2].set_title('Rotated image', y=-0.15, fontsize=14)
axs[2].axis('off')

# Crop the rotated image to match the original size
cropped_img = rotated_img[shape[0]//2:(shape[0]//2)*3, shape[1]//2:(shape[1]//2)*3]

# Display the cropped image
axs[3].imshow(cropped_img)
axs[3].set_title('Cropped image', y=-0.3, fontsize=14)
axs[3].axis('off')
```

```python
# Import necessary libraries for affine transformation
from skimage import transform
from skimage import img_as_float
import math
import numpy as np
import matplotlib.pyplot as plt

# Convert the image to floating point format for accurate transformations
img = img_as_float(image)

# Set up figure for displaying the transformed image
plt.figure(figsize=(6, 6))

# Define the affine transformation, which includes rotation, translation, scaling, and shearing
# - rotation: rotates the image by pi/6 radians (30 degrees)
# - translation: moves the image by 100 pixels along the x-axis and 40 pixels along the y-axis
# - scale: scales the image uniformly
# - shear: sets the shear transformation to 0 (no shear)
tf = transform.AffineTransform(rotation=np.pi / 6., translation=(100, 40), scale=0.25, shear=0)

# Apply backward mapping (inverse transform) to warp the image, ensuring smooth transformation without gaps
tf_img = transform.warp(img, tf.inverse)

# Display the transformed image
plt.imshow(tf_img, cmap='gray')
plt.axis('off')  # Hide axis for a cleaner display
```
